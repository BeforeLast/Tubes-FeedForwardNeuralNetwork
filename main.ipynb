# Train dengan 10-fold Cross Validation

```python
from random import randrange

#randomly split data
split = randrange(len(x))

x_train = x[0:split] 
x_test = x[split:]

y_train = y[0:split] 
y_test = y[split:]

print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
#because its 10 fold cross validation, we train 10 times using train = 0.9, validate = 0.1
#agar bisa dilihat lebih jelas utk semua konfigurasi, we train it 10 times
#its a bit bruteforce but yaudah lah ya

#config0
x_train0 = x_train[0:int(0.9*len(x_train))]
y_train0 = y_train[0:int(0.9*len(y_train))]
x_validate0 = x_train[int(0.9*len(x_train)):]
y_validate0 = y_train[int(0.9*len(y_train)):]

#config1
x_train1 = x_train[0:int(0.1*len(x_train))]
x_train1_join = x_train[int(0.2*len(x_train)):]
x_train1 = np.append(x_train1,x_train1_join, axis=0)
y_train1 = y_train[0:int(0.1*len(y_train))]
y_train1_join = y_train[int(0.2*len(y_train)):]
y_train1 = np.append(y_train1,y_train1_join, axis=0)
x_validate1 = x_train[int(0.1*len(x_train)):int(0.2*len(x_train))]
y_validate1 = y_train[int(0.1*len(y_train)):int(0.2*len(y_train))]

#config2
x_train2 = x_train[0:int(0.2*len(x_train))]
x_train2_join = x_train[int(0.3*len(x_train)):]
x_train2 = np.append(x_train2,x_train2_join, axis=0)
y_train2 = y_train[0:int(0.2*len(y_train))]
y_train2_join = y_train[int(0.3*len(y_train)):]
y_train2 = np.append(y_train2,y_train2_join, axis=0)
x_validate2 = x_train[int(0.2*len(x_train)):int(0.3*len(x_train))]
y_validate2 = y_train[int(0.2*len(y_train)):int(0.3*len(y_train))]

#config3
x_train3 = x_train[0:int(0.3*len(x_train))]
x_train3_join = x_train[int(0.4*len(x_train)):]
x_train3 = np.append(x_train3,x_train3_join, axis=0)
y_train3 = y_train[0:int(0.3*len(y_train))]
y_train3_join = y_train[int(0.4*len(y_train)):]
y_train3 = np.append(y_train3,y_train3_join, axis=0)
x_validate3 = x_train[int(0.3*len(x_train)):int(0.4*len(x_train))]
y_validate3 = y_train[int(0.3*len(y_train)):int(0.4*len(y_train))]

#config4
x_train4 = x_train[0:int(0.4*len(x_train))]
x_train4_join = x_train[int(0.5*len(x_train)):]
x_train4 = np.append(x_train4,x_train4_join, axis=0)
y_train4 = y_train[0:int(0.4*len(y_train))]
y_train4_join = y_train[int(0.5*len(y_train)):]
y_train4 = np.append(y_train4,y_train4_join, axis=0)
x_validate4 = x_train[int(0.4*len(x_train)):int(0.5*len(x_train))]
y_validate4 = y_train[int(0.4*len(y_train)):int(0.5*len(y_train))]

#config5
x_train5 = x_train[0:int(0.5*len(x_train))]
x_train5_join = x_train[int(0.6*len(x_train)):]
x_train5 = np.append(x_train5,x_train5_join, axis=0)
y_train5 = y_train[0:int(0.5*len(y_train))]
y_train5_join = y_train[int(0.6*len(y_train)):]
y_train5 = np.append(y_train5,y_train5_join, axis=0)
x_validate5 = x_train[int(0.5*len(x_train)):int(0.6*len(x_train))]
y_validate5 = y_train[int(0.5*len(y_train)):int(0.6*len(y_train))]

#config6
x_train6 = x_train[0:int(0.6*len(x_train))]
x_train6_join = x_train[int(0.7*len(x_train)):]
x_train6 = np.append(x_train6,x_train6_join, axis=0)
y_train6 = y_train[0:int(0.6*len(y_train))]
y_train6_join = y_train[int(0.7*len(y_train)):]
y_train6 = np.append(y_train6,y_train6_join, axis=0)
x_validate6 = x_train[int(0.6*len(x_train)):int(0.7*len(x_train))]
y_validate6 = y_train[int(0.6*len(y_train)):int(0.7*len(y_train))]

#config7
x_train7 = x_train[0:int(0.7*len(x_train))]
x_train7_join = x_train[int(0.8*len(x_train)):]
x_train7 = np.append(x_train7,x_train7_join, axis=0)
y_train7 = y_train[0:int(0.7*len(y_train))]
y_train7_join = y_train[int(0.8*len(y_train)):]
y_train7 = np.append(y_train7,y_train7_join, axis=0)
x_validate7 = x_train[int(0.7*len(x_train)):int(0.8*len(x_train))]
y_validate7 = y_train[int(0.7*len(y_train)):int(0.8*len(y_train))]

#config8
x_train8 = x_train[0:int(0.8*len(x_train))]
x_train8_join = x_train[int(0.9*len(x_train)):]
x_train8 = np.append(x_train8,x_train8_join, axis=0)
y_train8 = y_train[0:int(0.8*len(y_train))]
y_train8_join = y_train[int(0.9*len(y_train)):]
y_train8 = np.append(y_train8,y_train8_join, axis=0)
x_validate8 = x_train[int(0.8*len(x_train)):int(0.9*len(x_train))]
y_validate8 = y_train[int(0.8*len(y_train)):int(0.9*len(y_train))]

#config9
x_train9 = x_train[int(0.1*len(x_train)):]
y_train9 = y_train[int(0.1*len(x_train)):]
x_validate9 = x_train[0:int(0.1*len(x_train))]
y_validate9 = y_train[0:int(0.1*len(y_train))]
```

```python
#config 0
#redefine model
Model_Test0 = FFNN('./file/models/Iris_Model_Test.json')
Model_Test0.setLearningRate(0.01)
#train using training data
Model_Test0.train(x_train0, y_train0, 200, 1, 0.0000000000001)
#predict validate
mymodel_predict = Model_Test0.batch_predict(x_validate0)
#reformat answers
finalAns = []
for item in mymodel_predict:
    if item[0]>0.5:
        finalAns.append(1)
    else:
        finalAns.append(0)
print(finalAns)
#reformat y_validate
y_validate = [y_validate0[i][0] for i in range (len(y_validate0))]
print(y_validate)
confusionMatrix = algorithm.Util.findConfusionMatrix(y_validate, finalAns)
print("confusion matrix: ", end="\n")
print(confusionMatrix)
print("we have multiple classified data = setosa, versicolor, virginica : 0,1,2 <- in that order")
print("accuracy: " + str(algorithm.Util.getAccuracy(confusionMatrix)))
print("macro precision: " + str(algorithm.Util.getMacroPrecision(confusionMatrix)))
print("macro recall: " + str(algorithm.Util.getMacroRecall(confusionMatrix)))
print("macro f1 score: " + str(algorithm.Util.getMacroF1Score(confusionMatrix)))
```
